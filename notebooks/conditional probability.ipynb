## Bayes' theorem

$P(A | B) = \frac{P(A \cap B)}{P(B)}$

$P(B | A) = \frac{P(A \cap B)}{P(A)}$


$$P(A | B) = \frac{P(B | A) P(A)}{P(B)}$$

$P(A|B)$ ... conditional probability of  $A$ occurring given $B$ is true, "posterior"

$P(B|A)$ ... "likelihood" of $B$ occurring given $A$ is true

P(A) ... "prior"

P(B) ... "marginal probabilities" (integrated) of observing A,B (A and B) independently of each other, all possibilities of B can happen

## Covid test

test

- $99\%$ sensitivity (TP)
- $99\%$ specifity (TN)

Let the prevalance of the desease in population be $p=0.5\%$ ($0.5\%$ has the disease in the population

Q: What is the probability that a random individual has the disease if test is  positive?

$$P(A | B) = \frac{P(B | A) P(A)}{P(B)}$$

&nbsp;

$$P(sick|+) = \frac{P(+|sick)P(sick)}{P(+|sick) P(sick) + P(+|not\  sick)P(not\ sick)}$$

&nbsp;

$$= \frac{0.99 \times 0.005}{0.99 \times 0.005 + 0.01 \times 0.995} \approx 33.2$$

&nbsp;

If $1000$ individual are tested , with $p=0.005$ we expect $995$ $\ominus$ and $5$ $\oplus$.

Among $995$ the False Positive of the test $0.01$ gives $\doteq$ $10$ FP,

among $5$, the test accuracy of $0.99$ yields $\doteq$ $5$.

Therefore, in total we have 15 positive results and only 5 True Positive, which is $\approx 33\%$

Try for more realistic data for Czech republic and test accuracy... You can try to plot it.

```python
import numpy as np
import matplotlib.pyplot as plt

def f(p,t):
    return t*p/(t*p+(1.-t)*(1.-p))

p = np.linspace(0.,1,50)
t = np.linspace(0.,1.,50)

P,T = np.meshgrid(p,t)
Z = f(P,T)
plt.contourf(P,T,Z, 20)
plt.colorbar()
plt.xlabel("prevalence")
plt.ylabel("test accuracy");
```

```python
contours = plt.contour(P, T, Z, 20, colors="black", linewidths=1.)
plt.clabel(contours, inline=True, fontsize=7)

plt.imshow(Z, extent=[0, 1, 0, 1], origin='lower', alpha=0.5)
plt.colorbar()
plt.xlabel("prevalence")
plt.ylabel("test accuracy");
```
